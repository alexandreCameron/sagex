```python
car7 = pd.read_csv('car_2017_encoding_latin-1.csv', encoding='latin-1')
car8 = pd.read_csv('car_2018.csv')
claim7 = pd.read_csv('claim_2017.csv')
claim8 = pd.read_csv('claim_2018_delimiter_semicol.csv', sep=';')
cli7 = pd.read_csv('client_2017.csv')
cli8 = pd.read_csv('client_2018.csv')
```

#### Idem for the other tables:

```python
car8.drop_duplicates(inplace=True)
claim7.drop_duplicates(inplace=True)
claim8.drop_duplicates(inplace=True)
cli7.drop_duplicates(inplace=True)
cli8.drop_duplicates(inplace=True)
```

#### Now we can merge tables for each year:

```python
cli_car7 = pd.merge(cli7, car7, on='ID', how='outer')
data2017 = pd.merge(cli_car7, claim7, on='ID', how='outer')
```

```python
cli_car8 = pd.merge(cli8, car8, on='ID', how='outer')
data2018 = pd.merge(cli_car8, claim8, on='ID', how='outer')
data2018.head()
```
